```python
# Dependencies and Setup
import pandas as pd
import numpy as np
from IPython.core.display import HTML

# Raw data file
file_to_load = "purchase_data-Copy1.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.rename(index={0:"",1:"",2:"",3:"",4:"",},inplace = True)
purchase_data1 = purchase_data
purchase_data2 = purchase_data
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
#count unique player identifiers from SN column
total_players = purchase_data.SN.nunique()
#make into df with desired heading 
total_players_table = pd.DataFrame({"Total Players": [total_players]})
#show 
total_players_table.rename(index={0:""},inplace = True)

total_players_table.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#find the mean of all purchase prices logged and store as variable
average_price = purchase_data["Price"].mean()
#count the total number of purchases (every time a price was logged) and store as variable
number_of_purchases = purchase_data["Price"].count()
#count the number of unique item ids and store as variable
unique_items = purchase_data["Item ID"].nunique()
#calculate total revenue (sum total of all items priced as sold) and store as variable
total_revenue = purchase_data["Price"].sum()
#create summary table of variables with desired names
summary_table = pd.DataFrame({"Average Price": [average_price],
                              "Number of Purchases": [number_of_purchases],
                              "Number of Unique Items": [unique_items],
                              "Total Revenue": [total_revenue]})
#show
summary_table.rename(index={0:""},inplace = True)
summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#make a new df with all unique players to avoid overcounting
demo_df = purchase_data.drop_duplicates(subset="SN", keep='first', inplace=False)
#make alphabetical to checck
demo_check = demo_df.sort_values('SN')
#demo_check
number_of_males = demo_check.Gender.value_counts()['Male']
#number_of_males
number_of_females = demo_check.Gender.value_counts()['Female']
#number_of_females
number_of_other = demo_check.Gender.value_counts()['Other / Non-Disclosed']
#number_of_other
male_percent = (number_of_males / total_players) * 100
#male_percent
female_percent = (number_of_females / total_players) * 100
#female_percent
other_percent = (number_of_other / total_players) * 100
#other_percent
demo_df = pd.DataFrame({
    'Number': [number_of_males, number_of_females, number_of_other],
    'Percent': [male_percent,female_percent, other_percent],
})
#demo_df- check values
demo_df.rename(index={0:'Male',1:'Female',2:'Other/Non-Discolsed'}, inplace=True)
demo_df
```

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
male_purchases_df = purchase_data.loc[purchase_data['Gender'] == "Male"]
#male_purchases.head()
mp = male_purchases_df.Gender.value_counts()
mp = mp.values
ppm = mp / number_of_males
female_purchases_df = purchase_data.loc[purchase_data['Gender'] == "Female"]
fp = female_purchases_df.Gender.value_counts()
fp = fp.values
ppf = fp / number_of_females
#female_purchases
other_purchases_df = purchase_data.loc[purchase_data['Gender'] == "Other / Non-Disclosed"]
op = other_purchases_df.Gender.value_counts()
op = op.values
ppo = op /number_of_other
#other_purchases
male_mean_price = male_purchases_df["Price"].mean()
#male_mean_price
female_mean_price = female_purchases_df["Price"].mean()
#female_mean_price
other_mean_price = other_purchases_df["Price"].mean()
#other_mean_price
demo_sales_df = pd.DataFrame({
    'Total Purchase Count by Gender': [mp, fp, op],
    'Avg. Purchase Price by Gender': [male_mean_price,female_mean_price, other_mean_price],
    'Average Purchase Count by Gender': [ppm, ppf, ppo],
})
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python


# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
age_bin_names = ["<10", "10-14", "15-19", "20-24", "25-29",
                "30-34", "35-39", "40+"]
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins,
                                        labels=age_bin_names)

agegroups = purchase_data.groupby('Age Group').nunique()
#agegroups
agecount_df  = agegroups.filter(['SN'], axis=1)

agecount_df.rename({'SN':'Unique Players'}, axis=1)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Most Popular Items
# Identify the 5 most popular items by purchase count, then list (in a table):
# Item ID
# Item Name
# Purchase Count

# Item Price
# Total PurchaseItem IDValue

popitems = purchase_data2.groupby("Item ID").count().sort_values(["Price"], ascending=False).reset_index()
popitems["Item Name"] = purchase_data2.groupby("Item Name").count().sort_values(["Price"], ascending=False).reset_index()
popitems["Price"] = purchase_data2.groupby("Price").count().sort_values(["Item ID"], ascending=False).reset_index()
popitems["Total Sales Per Item"] = popitems.Age * popitems.Price
popcut = popitems.drop(columns=['Gender','SN','Age','Age Group'])
popcut.rename({'Purchase ID':'Purchase Count'}, axis=1, inplace=True)
popcut = popcut[:5]
popcut
```

## Most Profitable Items

* Display a preview of the data frame



```python
profitems = purchase_data2.groupby("Item ID").count().sort_values(["Price"], ascending=False).reset_index()
profitems["Item Name"] = purchase_data2.groupby("Item Name").count().sort_values(["Price"], ascending=False).reset_index()
profitems["Price"] = purchase_data2.groupby("Price").count().sort_values(["Item ID"], ascending=False).reset_index()
profitems["Total Sales Per Item"] = profitems.Age * profitems.Price
profcut = profitems.drop(columns=['Gender','SN','Age','Age Group'])
profcut.rename({'Purchase ID':'Purchase Count'}, axis=1, inplace=True)
profcut.sort_values(by=['Total Sales Per Item'], ascending=False)
profcut = profcut[:5]
profcut
```
